Вообще говоря, может быть как типа object (текстовые), так и int (численные). В случае int можно считать, что категории закодированы числами. - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html#pandas.DataFrame.dropna

```python
tmp_df = df.copy()
categorical = tmp_df.nunique()
categorical = categorical[categorical < 100]
categorical = tmp_df[categorical.index].select_dtypes(include=['object']).columns
non_categorical = tmp_df.nunique()
non_categorical = non_categorical[non_categorical >= 100]
non_categorical = tmp_df[non_categorical.index].select_dtypes(include=['object']).columns
tmp_df.dropna(subset=categorical, inplace=True)
tmp_df[non_categorical] = tmp_df[non_categorical].fillna('')
tmp_df.shape
```

Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.
