```python
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
sliced = df[['publishedAt', 'title']].copy()
sliced.publishedAt = sliced.publishedAt.dt.floor('d')

unique = sliced.drop_duplicates()

fig, axis = plt.subplots()
plt.title('   ')
plt.xlabel('')
plt.ylabel('')
plt.plot(unique.groupby('publishedAt').count(), 'o--', ms=4, lw=0.8)
ticks = unique.publishedAt.drop_duplicates().sort_values()[::2]
axis.set_xticks(ticks)
_ = axis.set_xticklabels(ticks.map(lambda x: f'{str(x)[5:10].replace('-', '.')}'), rotation=45)
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
sliced = df[['country', 'title']].copy()

unique = sliced.drop_duplicates()
assert len(sliced) == len(unique)

fig, axis = plt.subplots()
plt.title('    ')
plt.xlabel('')
plt.ylabel('')
values = unique.groupby('country').count().title
_ = plt.bar(values.index, values)
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
from matplotlib.ticker import ScalarFormatter

formatter = ScalarFormatter(useMathText=True)
formatter.set_powerlimits((0, 0)) 
```

```python
sliced = df[(df.country == 'FR') & (df.comments_disabled == False)][['title', 'comment_count', 'view_count']].copy()

unique = sliced.drop_duplicates()
assert len(sliced) == len(unique)
data = sliced.sort_values(['view_count', 'comment_count'], ascending=True)[['view_count', 'comment_count']]

fig, axis = plt.subplots()
axis.set_title('      ')
axis.set_xlabel(' ')
axis.set_ylabel(' ')
axis.xaxis.set_major_formatter(formatter)

_ = axis.scatter(data.view_count, data.comment_count)
```

```python
fig, axis = plt.subplots()
axis.set_title('      ')
axis.set_xlabel(' ')
axis.set_ylabel(' ')
axis.xaxis.set_major_formatter(formatter)

_ = axis.scatter(data.view_count[:-6], data.comment_count[:-6])
```

```python
capped = data[data.view_count <= 10 ** 6]

fig, axis = plt.subplots()
axis.set_title('      ')
axis.set_xlabel(' ')
axis.set_ylabel(' ')
axis.xaxis.set_major_formatter(formatter)

_ = axis.scatter(capped.comment_count, capped.view_count)
```
